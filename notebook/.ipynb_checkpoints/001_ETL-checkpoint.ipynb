```python
import pandas as pd
from pandas import DataFrame
import glob as glob
from datetime import datetime
import chardet
import re
import os
```

```python
def elimina_archivo(f):
   """
         Elimina los archivos *_mod.dat que se generaron con la información pre-procesada
    """        
    if os.path.exists(f):
        try:
            # Paso 2: Intentar eliminar el archivo
            os.remove(f)
            print(f"El archivo '{f}' ha sido eliminado correctamente.")
        except PermissionError:
            print(f"Error: No tienes permisos para eliminar el archivo '{f}'.")
        except Exception as e:
            print(f"Ocurrió un error inesperado al eliminar el archivo: {e}")
    else:
        print(f"El archivo '{f}' no existe.")
```

```python
#Busca los archivos generados con el pre-procesamiento y los elimina
files = glob.glob("../data/001_raw/*/*_mod.dat")
print(len(files))
#files
lista_dfs = [elimina_archivo(file) for file in files]
```

```python
def encoding_f(file):
   """
         Encontrar el decoding de los archivos
         Retorna el decoding
    """        
    with open(file, 'rb') as f:
        resultado = chardet.detect(f.read())
        
    codificacion = resultado['encoding']
    return codificacion        
```

```python
def pre_procesar_archivo(f, type_encoding):
    """
    Elimina renglones vacíos o con solo espacios/tabs de un archivo.
    Elimina mas de un tab y solo conserva uno
    Verifica si la segunda columna no es entero elimina ese dato
    Args:
        nombre_archivo_entrada (f): El nombre del archivo a leer.
        type_encoding : (Tipo de encoding del archivo
    """
    #f_salida = f.replace("001_raw","002_processed")
    f_salida = f
    f_salida_tmp = f
    f_salida = f_salida.replace(".dat","_mod.dat")    
    #f_salida = f_salida.replace("001_raw","002_processed")
    f_salida_tmp = f_salida_tmp.replace("001_raw","002_processed")
    #print("Archivo:", f_salida)
    #print("Archivo_tmp:", f_salida_tmp)
    try:
        with open(f, 'r', encoding=type_encoding) as f_entrada:
            #contenido_modificado = re.sub(r'\t+', '\t', contenido)
            lineas_limpias = []
            i=0
            for linea in f_entrada:
                # Usa strip() para eliminar espacios/tabs/saltos de línea del inicio y fin
                # Si la línea resultante no está vacía, la añadimos a la lista
                #print(linea)
                #Salta las lineas 0, 2 y 3
                if (i!= 0 and i !=2 and i!=3):
                    tab_count = linea.count('\t')
                    linea_sin_tabs = linea
                    #Si el número de tabs > 0 entonces el archivo está separado por tabs
                    if tab_count > 1:
                        linea_sin_tabs = re.sub(r'\t+', '\t', linea)              # Solo deja un tab
                    if (i == 1):                                                  # Si es el encabezado no verifica la segunda columna donde se detectó un 
                        lineas_limpias.append(re.sub('\t', ',', linea_sin_tabs))  #error, aparece otra fecha en lugar del índice
                    elif linea_sin_tabs.strip():                                  # Verifica que la línea no sea vacia, si es así, se la salta
                        datos = linea_sin_tabs.split('\t')
#                        print(datos)
#                        print(len(datos))
                        #El archivo está separado por tabs
                        if len(datos) >2:
                            es_numero = datos[1].isdigit()                            
                            if es_numero:
                                #lineas_limpias.append(linea_sin_tabs)
                                lineas_limpias.append(re.sub('\t', ',', linea_sin_tabs))
                            #else:
                            #    del datos[1]
                            #    if(not datos[1].isdigit()):
                            #        del datos[1]
                                #El archivo ahora se separa por comas
                            #    linea_modificada = ','.join(datos)
                            #    lineas_limpias.append(linea_modificada)
                        #El archivo está separado por comas
                        else:
                            lineas_limpias.append(linea_sin_tabs)                
                i = i+1
        #Se imprime el archivo procesado              
        with open(f_salida, 'w', encoding=type_encoding) as f_salida:
            # Escribe las líneas limpias en el nuevo archivo
            f_salida.writelines(lineas_limpias)
            print(f"Archivo Procesado guardado en '{f_salida}'")
    except FileNotFoundError:
        print(f"Error: El archivo '{f_salida}' no fue encontrado.'{f}'")
    except Exception as e:
        print(f"Ocurrió un error: {e}")
```

```python
#Genera los archivos preprocesados
files = glob.glob("../data/001_raw/*/*ESOLMET*10*.dat") + glob.glob("../data/001_raw/*/*ESOLMET*diez*.dat")
print(len(files))
i =1
for file in files:
    print(i)
    type_encoding = encoding_f(file)
    pre_procesar_archivo(file, type_encoding)    
    i = i+1
```

```python
def importa_excel(f)-> DataFrame:
    """
            Importa el archivo    
        Args:
            nombre_archivo_entrada (f): El nombre del archivo a leer.
        Retorna el dataframe con la información del archivo    
    """

    custom_date_parser = lambda x: datetime.strptime(x, '%d-%m-%Y %H:%M')
    custom_date_parser_2 = lambda x: datetime.strptime(x, '%Y-%m-%d %H:%M:%S')

    print("Archivo a procesar:",f)
    type_encoding = encoding_f(f)
    try:        
        tmx = pd.read_csv(f,encoding=type_encoding,sep=",",index_col=0, parse_dates=True)#, date_parser=custom_date_parser_2)        
    except Exception as e:  
        #El archivo no está separado por comas, probamos con separadores de espacio        
        try:
            tmx = pd.read_csv(f,encoding=type_encoding,sep=',',index_col=0, parse_dates=True, dayfirst=True, date_parser=custom_date_parser)
            print("Se formateó la fecha")
        except Exception as e:
            print("Error al procesar el archivo:",{e})
        
    tmx = tmx[tmx.index.year>=2020]
    return tmx
```

```python
#Procesa todos los archivos pre-procesados y los concatena en un dataframe
files = glob.glob("../data/001_raw/*/*_mod.dat")
print(len(files))
#files
lista_dfs = [importa_excel(file) for file in files]
tmx = pd.concat(lista_dfs)
#lista_dfs
```
